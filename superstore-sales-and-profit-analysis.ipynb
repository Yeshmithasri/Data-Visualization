```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

ss = pd.read_csv("C:/Users/acer/Downloads/Sample - Superstore.csv")

ss.head(10)
```

```python
ss.describe()
```

```python
ss.shape
```

```python
ss['Order Date'] = pd.to_datetime(ss['Order Date'], errors = 'coerce')
ss['Ship Date'] = pd.to_datetime(ss['Ship Date'], errors = 'coerce') 

ss['Order Month'] = ss['Order Date'].dt.month 
ss['Order Year'] = ss['Order Date'].dt.year
ss['Order Day of Week'] = ss['Order Date'].dt.dayofweek

ss.head()
```

```python
unique_years = ss['Order Date'].dt.year.unique()

unique_years
```

```python
# Filter data for the year 2014
sales14 = ss[ss['Order Date'].dt.year == 2014]

# Group by month and calculate the total sales
monthsales14 = sales14.groupby(sales14['Order Date'].dt.month)['Sales'].sum().reset_index()

# Sort the months
monthsales14 = monthsales14.sort_values(by='Order Date')

# Plotting the line graph
plt.figure(figsize=(8, 6))
plt.plot(monthsales14['Order Date'], monthsales14['Sales'], marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Monthly Sales in 2014')
plt.xticks(monthsales14['Order Date'], ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```
